First of all I have to Import the basic libraries for plotting in python.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Then I have to import a DataSet:

```python
kashti = sns.load_dataset('titanic')
kashti.head()
```

```python
kashti.isnull().sum()
```

```python
#Bar plot
kashti.survived.value_counts().plot(kind='bar')  # here value_counts() is function of pandas which finds the 
                                                # unique value of survived as 0 and 1(true or false)
```

```python
                                                    #countplot always show in bar plot style
sns.countplot(x='survived',hue='sex',data=kashti)   #it is inside seaborn library 
```

```python
sns.barplot(x='survived',y='fare',data=kashti,hue='embark_town')  
```

```python
# scatter plot with seaborn
sns.scatterplot(x='age',y='fare',data=kashti,hue='pclass')
```

```python
# scatter plot( is me ye khiyal rakhna hota hai ke dono x aur y ki value numeric ho)
plt.scatter('age','fare',data=kashti)
```

```python
sns.pairplot(kashti)
```

```python
kashti.hist()
```

```python
kashti.hist(bins=10,figsize=(7,10))
```

```python
kashti.hist(bins=10,figsize=(7,7),grid=False)
```

```python
kashti.columns.unique()
```

```python
sns.boxplot(data=kashti,x='who',y='age',hue='pclass')
```

```python
sns.boxenplot(data=kashti,x='who',y='age',hue='embarked')
```

```python
g=sns.FacetGrid(kashti,hue='survived',col='embark_town',margin_titles= True,palette={0:'red',1:'blue'})
g=g.map(plt.scatter,'fare','age',edgecolor='w').add_legend()
```

```python
sns.stripplot(kashti,x='sex',y='age',hue='survived')
```

```python
sns.stripplot(kashti,x='sex',y='age',hue='who')
```
