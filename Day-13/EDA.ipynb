```python
# swarmplot
sns.swarmplot(df , x = 'survived', y = 'age', hue = 'who')
```

```python
# relation of two numeric variables 
import seaborn as sns

sns.lmplot(data=df, x='age', y='fare', hue='who', fit_reg=False, markers=['o', 's', 'D']) # fit_reg is for removing
                                                                                        #the outlier line in the plot
```
